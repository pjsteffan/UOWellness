```python
import os
from docx import Document
from openai import OpenAI
from dotenv import load_dotenv
import yaml
import pypandoc

pypandoc.download_pandoc()


load_dotenv()
client = OpenAI()
```

```python
import os

def get_file_names_without_extension(folder_path):
    """
    Get a set of file names without extensions from the given folder.
    
    Args:
    folder_path (str): Path to the folder.

    Returns:
    set: Set of file names without extensions.
    """
    return {os.path.splitext(file)[0] for file in os.listdir(folder_path)}

def find_unique_files(folder1, folder2):
    """
    Find files in folder1 that are not in folder2 based on file names without considering file extensions.

    Args:
    folder1 (str): Path to the first folder.
    folder2 (str): Path to the second folder.

    Returns:
    list: List of file names in folder1 not present in folder2.
    """
    folder1_files = get_file_names_without_extension(folder1)
    folder2_files = get_file_names_without_extension(folder2)

    unique_files = folder1_files - folder2_files
    return list(unique_files)




def extract_text_from_docx(docx_path):
    """
    Extracts text from a .docx file.
    """
    output = pypandoc.convert_file(docx_path, 'plain', outputfile="somefile.txt")

def iterate_through_folder(folder_path,context):
    """
    Iterates through all .docx files in the specified folder and extracts text from each.
    """
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.endswith(".docx"):
                file_path = os.path.join(root, file)
                print(file_path)
                output = pypandoc.convert_file(file_path, 'plain', outputfile=os.path.join('/app/text_files', os.path.basename(file_path.replace('.docx','.txt'))))
                assert output == ""
                
                with open(os.path.join('/app/text_files', os.path.basename(file_path.replace('.docx','.txt')))) as f:
                    text = f.read() 
                    fText = call_llm(context,text)
                    print(fText)

                    # Create the output file path by combining the output folder and filename
                    output_filepath = os.path.join('/app/formatted_text', os.path.basename(file_path.replace('.docx','.txt')))
                    
                    with open(output_filepath, "w") as g:
                        g.write(fText)
                    

def call_llm(context,message):
    completion = client.chat.completions.create(
    model="gpt-4o-mini",#gpt-3.5-turbo-0125,gpt-4-1106-preview
    messages=[
        {"role": "system", "content": context},
        {"role": "user", "content": message}
    ],
    temperature=0.6,
    )
    return completion.choices[0].message.content

```

```python
folder_path = '/app/WELLNESS_HTML_DOCS'

with open('/app/Prompting.yaml') as file:
    prompts = yaml.load(file, Loader=yaml.FullLoader)

context = prompts['context']

iterate_through_folder(folder_path,context)
```
